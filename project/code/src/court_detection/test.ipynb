```python
# Calude generated
# Visual test for CourtDataset
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from dataloader import CourtDataset

dataset = CourtDataset(type="train")
loader = DataLoader(dataset, batch_size=4, shuffle=False, num_workers=2)

# Get one batch
imgs, heatmaps_list, heatmaps_center = next(iter(loader))

print(f"Dataset size: {len(dataset)}")
print(f"Input batch shape: {imgs.shape}")           # (B, 3, H, W)
print(f"Number of kp heatmaps: {len(heatmaps_list)}")  # 14
print(f"One kp heatmap shape: {heatmaps_list[0].shape}")  # (B, 1, H, W)
print(f"Center heatmap shape: {heatmaps_center.shape}")   # (B, 1, H, W)

# ── Show the 4 images ───────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 4, figsize=(20, 4))
fig.suptitle("Input images", fontsize=14)

for i in range(4):
    frame = imgs[i].permute(1, 2, 0) * 0.5 + 0.5   # denormalise → [0,1]
    axes[i].imshow(frame.numpy())
    axes[i].axis("off")
    axes[i].set_title(f"Sample {i}")

plt.tight_layout()
plt.show()

# ── Show all 14 keypoint heatmaps for sample 0 ──────────────────────────────
fig, axes = plt.subplots(2, 7, figsize=(24, 6))
fig.suptitle("14 keypoint heatmaps — Sample 0", fontsize=14)

for kp_idx in range(14):
    row, col = divmod(kp_idx, 7)
    axes[row, col].imshow(heatmaps_list[kp_idx][0].squeeze(0).numpy(), cmap="hot")
    axes[row, col].axis("off")
    axes[row, col].set_title(f"KP {kp_idx}")

plt.tight_layout()
plt.show()

# ── Show the centre heatmap for all 4 samples ───────────────────────────────
fig, axes = plt.subplots(1, 4, figsize=(20, 4))
fig.suptitle("Court centre heatmap", fontsize=14)

for i in range(4):
    axes[i].imshow(heatmaps_center[i].squeeze(0).numpy(), cmap="hot")
    axes[i].axis("off")
    axes[i].set_title(f"Sample {i}")

plt.tight_layout()
plt.show()

# ── Overlay: image + all keypoints on sample 0 ──────────────────────────────
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
fig.suptitle("Image + keypoint peaks overlay — Sample 0", fontsize=14)

frame = imgs[0].permute(1, 2, 0) * 0.5 + 0.5
ax.imshow(frame.numpy())

for kp_idx in range(14):
    hm = heatmaps_list[kp_idx][0].squeeze(0).numpy()   # (H, W)
    peak = hm.argmax()
    py, px = divmod(peak, hm.shape[1])
    ax.scatter(px, py, s=40, zorder=5)
    ax.text(px + 3, py - 3, str(kp_idx), fontsize=7, color="white")

# Also plot the centre
hm_c = heatmaps_center[0].squeeze(0).numpy()
peak_c = hm_c.argmax()
py_c, px_c = divmod(peak_c, hm_c.shape[1])
ax.scatter(px_c, py_c, s=60, marker="*", color="red", zorder=6, label="centre")
ax.legend()
ax.axis("off")

plt.tight_layout()
plt.show()
```
